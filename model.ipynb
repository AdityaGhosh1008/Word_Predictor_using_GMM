```python
def extract_mfcc(audio, sample_rate,max_frames,n_mfcc=13):
    audio = audio.astype(np.float32)
    mfccs = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=n_mfcc)
    delta_mfccs = librosa.feature.delta(mfccs)

    # Compute Delta-Delta MFCCs
    delta2_mfccs = librosa.feature.delta(mfccs, order=2)

    # Combine MFCC, Delta MFCC, and Delta-Delta MFCC arrays in a single line
    combined_features = np.concatenate((mfccs, delta_mfccs))
    combined_features = np.concatenate((combined_features, delta2_mfccs))
    current_frames = combined_features.shape[1]

    # Pad or truncate to ensure a fixed number of frames

    return combined_features.transpose()
```

```python
def mfccstack(class_folder):
    mfccs_stack = np.empty((0, 39))  # Initialize an empty array with the desired number of columns (num_mfcc)

    files = os.listdir(class_folder)
    for filename in files:
        audio_path = os.path.join(class_folder, filename)
        sample_rate, audio = wavfile.read(audio_path)
        mfccs = extract_mfcc(audio, sample_rate, max_frames=100)
        mfccs_stack = np.vstack((mfccs_stack, mfccs))

    print(mfccs_stack.shape)
    print(mfccs_stack)
    return mfccs_stack

```

```python
import csv

test_path = "C:\\Users\\ghosh\\Downloads\\test.csv"

# Initialize an empty dictionary to store the data
csv_dict = {}
data_path = "C:\\Users\\ghosh\\Downloads\\SpeechCommandTest-20231114T162637Z-001\\SpeechCommandTest"
ans_dict={}
# Open the CSV file and read its contents
with open(test_path, 'r') as file:
    # Create a CSV reader
    csv_reader = csv.reader(file)

    # Read the header (first row) to skip it
    next(csv_reader)

    # Read the data and populate the dictionary
    for index, row in enumerate(csv_reader):
        # Use the first column as keys and the second column as values
        key = row[1]
        value = index

        # Add the data to the dictionary
        csv_dict[key] = value
        file_path = os.path.join(data_path, key)
        sample_rate, audio = wavfile.read(file_path)
        audio = audio.astype(np.float32)
        mfccs = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=13)
        delta_mfccs = librosa.feature.delta(mfccs)

        delta2_mfccs = librosa.feature.delta(mfccs, order=2)

        combined_features = np.concatenate((mfccs, delta_mfccs))
        combined_features = np.concatenate((combined_features, delta2_mfccs))
        features=combined_features.transpose()
        problist=[]
        for i in range(len(gmmlist)):
            prob = gmmlist[i].score(features)
            problist.append(prob)
        ans_dict[key]=problist.index(max(problist))
    print(ans_dict)

# Print the resulting dictionary

```

```python
# Write ans_dict into a CSV file
output_csv_path = "C:\\Users\\ghosh\\Downloads\\output.csv"
with open(output_csv_path, 'w', newline='') as output_file:
    csv_writer = csv.writer(output_file)
    
    # Write header
    csv_writer.writerow(['ID', 'Target'])  # Adjust column names as needed

    # Write data
    for key, value in ans_dict.items():
        csv_writer.writerow([csv_dict[key],value])
```
